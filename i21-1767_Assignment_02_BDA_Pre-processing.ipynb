```python
df1 = df1[['ARTICLE_ID', 'SECTION_TEXT']]
df1
```

```python
df = df1[0:50000]
df
```

```python
def clean_text(text):
    if isinstance(text, str):
        cleaned_text = re.sub(r'\s+', ' ',text.replace('\n', ' ')).strip()
        cleaned_text = re.sub(r'[^a-zA-Z\s]', '',cleaned_text)
        cleaned_text = cleaned_text.lower()
        return cleaned_text
    else:
        return text 

df['SECTION_TEXT'] = df['SECTION_TEXT'].apply(clean_text)
```

```python
df2 = df[0:5000]
```

```python
df2
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
df2.dropna(inplace=True)

def preprocess_text(text):
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not token.is_punct and not token.is_space]
    return tokens

df2['SECTION_TEXT'] = df2['SECTION_TEXT'].apply(preprocess_text)
```

```python
df2
```

```python
ind = 0
with open('result.txt', 'w') as f:
    for index, row in df2.iterrows():
        f.write(f'{row["ARTICLE_ID"]}, ')
        ind2 = row["ARTICLE_ID"]
        if ind2 == ind:
            f.write(' '.join(row["SECTION_TEXT"]))
        elif ind2!=ind:
            ind = row["ARTICLE_ID"]
            f.write(f'\n')
```
